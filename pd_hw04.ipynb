```python
import numpy as np
from IPython.display import display
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\PC\Desktop\AI Course - E-Com\files\mobile.csv")
df
```

```python
# 1. 
df['First Name'] = df.apply(lambda row: row['Customer Name'].split()[0], axis=1)
```

```python
# 2. 
df['Phone Company'] = df.apply(lambda row: row['Phone Number'].split("-")[0], axis=1).replace({'050': 'Pelephone', '051': 'Golan Telecom', '052': 'Cellcom', '053': 'Motorola', '054': 'Partner'})
df
```

```python
# 3. 
# a. 
display(df['Phone Company'].value_counts())

# b. 
display(df['Phone Company'].unique())

# c. 
display(df['Phone Company'].nunique())
```

```python
# 4. 
df['User Type'] = df.apply(lambda row: 'Talker' if row['Talked'] > row['Internet'] else 'Surfer', axis=1)
df['User Type vector'] = np.vectorize(lambda talk, internet: 'Talker' if talk > internet else 'Surfer')(df['Talked'], df['Internet'])
df
```

```python
# 5. 
display(df.sort_values('Talked'))
display(df.sort_values(['Talked', 'Internet']))
```

```python
# 6. 
display(df.nlargest(5, 'Talked'))
df.nsmallest(5, 'Internet')
```

```python
# 7. 
display(df[df.duplicated()])
df.drop_duplicates()
```
