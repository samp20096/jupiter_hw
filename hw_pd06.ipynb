```python
import numpy as np
from IPython.display import display
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\PC\Desktop\AI Course - E-Com\files\sport_athletes.csv")
df
```

```python
# 1. 
df.groupby('Sport Type').agg({'Weight(kg)': 'min', 'Height(cm)': 'max', 'Wins': 'max'}).rename(columns={'Weight(kg)': 'max_weight', 'Height(cm)': 'max_height', 'Wins': 'max_wins'})
```

```python
# 2. 
df.groupby('Sport Type')['Height(cm)'].nlargest(2)
```

```python
# 3.a. 
df.groupby(['Country', 'Sport Type']).agg({'Weight(kg)': 'max', 'Height(cm)': 'max', 'Wins': 'max'})
```

```python
# 3.b. 
df.groupby(['Country', 'Sport Type']).agg({'Weight(kg)': 'max', 'Height(cm)': 'max', 'Wins': 'max'}).loc['USA','Basketball']
```

```python
# 3.c. 
df.groupby(['Country', 'Sport Type']).agg({'Weight(kg)': 'max', 'Height(cm)': 'max', 'Wins': 'max'}).xs(level='Sport Type', key='Soccer')
```

```python
# 3.d. 
df[df['Sport Type'].isin(['Tennis', 'Soccer'])].groupby('Sport Type').max()
```

```python
# 3.e. 
df.groupby(['Country', 'Sport Type']).max().swaplevel().sort_index(level=['Sport Type', 'Country'])
```

```python
# 3.f. 
df.agg({'Height(cm)': ['max', 'min', 'mean', 'std'], 'Wins': ['max', 'min', 'mean', 'std']})
```

```python
# 4. 
new_rows = pd.DataFrame({ 'Country': ['Canada', 'USA'], 'Sport Type': ['Hockey', 'Skiing'],
'Athlete Name': ['Wayne Gretzky', 'Lindsey Vonn'], 'Height(cm)': [183, 172],
'Weight(kg)': [84, 68], 'Wins': [3, 4], 'Olympic Date': ['01/02/2002', '12/02/2010'] })

pd.concat([df, new_rows])
```
