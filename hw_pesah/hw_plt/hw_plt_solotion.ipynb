```python
import numpy as np
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
```

# תרגיל 1: גרף קו בסיסי (Line Chart)

הוראות:
צור גרף קו שמייצג את הגובה של ילד לאורך השנים.
השתמש ברשימות הבאות:

years = [2015, 2016, 2017, 2018, 2019]
height = [110, 115, 120, 125, 132]

הוסף:

כותרת לגרף

תוויות לצירים

```python
years = [2015, 2016, 2017, 2018, 2019]
height = [110, 115, 120, 125, 132]

plt.plot(years, height)
plt.xlabel('Year')
plt.ylabel('Height')
plt.title('Height per Year')
plt.xticks(years)
plt.show()
```

 # תרגיל 2: Subplot של טמפרטורה וגשם
הוראות:

צור Subplot עם שני גרפים בשורה אחת:

גרף קו המציג את הטמפרטורות לפי חודשים

גרף עמודות המציג את כמות הגשמים לפי חודשים

```python
months = ["Jan", "Feb", "Mar", "Apr", "May"]
temperature = [12, 14, 17, 21, 25]
rain = [80, 60, 55, 40, 20]

plt.subplot(1, 2, 1)
plt.title('Temperature per Month')
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.plot(months, temperature)

plt.subplot(1, 2, 2)
plt.title('Rain per Month')
plt.xlabel('Month')
plt.ylabel('Rain')
plt.plot(months, rain)

plt.show()
```

# 🟥 תרגיל 3: היסטוגרמה (Histogram)
הוראות:

צור היסטוגרמה שמראה את ציוני התלמידים במבחן.

בחר bins=5 או bins=10 כדי לראות איך מספר הקבוצות משפיע על תצוגת הנתונים.



```python
grades = [70, 85, 90, 76, 65, 95, 88, 73, 80, 78, 67, 92, 60, 83, 89, 55, 91, 74, 69, 87, 81, 79, 62, 77, 93]

plt.hist(grades, bins=10)
plt.title('Amount of students per range of grades')
plt.xlabel('Grades')
plt.ylabel('Amount')
plt.show()
```

 # תרגיל 4: גרף עמודות (Bar Chart)
הוראות:

צור גרף עמודות שמשווה בין מספר תלמידים בכל שכבת גיל:

```python
age_groups = ['10-12', '13-15', '16-18', '19-21']
students_count = [30, 45, 25, 10]

plt.bar(age_groups, students_count)
plt.title('Number of Students by Age Group')
plt.xlabel('Age Groups')
plt.ylabel('Amount of Students')
plt.show()
```
