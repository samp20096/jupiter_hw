```python
import numpy as np
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
```

```python
students = pd.read_csv(r"C:\Users\PC\Desktop\AI Course - E-Com\files\hw_pesah\students.csv")
teachers = pd.read_csv(r"C:\Users\PC\Desktop\AI Course - E-Com\files\hw_pesah\teachers.csv")
courses = pd.read_csv(r"C:\Users\PC\Desktop\AI Course - E-Com\files\hw_pesah\courses.csv")
students_courses = pd.read_csv(r"C:\Users\PC\Desktop\AI Course - E-Com\files\hw_pesah\studnet_course.csv")

display("Students", students)
display("Teachers", teachers)
display("Courses", courses)
display("Students - Courses", students_courses)
```

# ניקוי מידע חסר
מחק את כל הרשומות המכילות מידע חסר

מחק את כל הרשומות הכפולות

```python
students.dropna(inplace=True)
teachers.dropna(inplace=True)
courses.dropna(inplace=True)
students_courses.dropna(inplace=True)

display("Students", students)
display("Teachers", teachers)
display("Courses", courses)
display("Students - Courses", students_courses)
```

```python
students.drop_duplicates(inplace=True)
teachers.drop_duplicates(inplace=True)
courses.drop_duplicates(inplace=True)
students_courses.drop_duplicates(inplace=True)

display("Students", students)
display("Teachers", teachers)
display("Courses", courses)
display("Students - Courses", students_courses)
```

# שאלה 1
ספור כמה תלמידים יש מכל מגדר בטבלת students.csv

הצג גם גרף עמודות (bar plot)

```python
gender_count = students['gender'].value_counts()
display("Gender Count", gender_count)

plt.bar(gender_count.index, gender_count.values)
plt.xlabel("Gender")
plt.ylabel("Amount")
plt.title("Gender Count")
plt.show()
```

# שאלה 2
צור היסטוגרמה להצגת התפלגות הגילאים של התלמידים

```python
plt.hist(students['age'], bins=range(int(students['age'].min()), int(students['age'].max()) + 1))
plt.xlabel("Age")
plt.ylabel("Amount")
plt.title("Count Per Age")
plt.show()
```

# שאלה 3
ספר כמה קורסים יש בכל קטגוריה בטבלת courses.csv

הצג גם גרף עמודות (bar plot)



```python
amount_courses = courses.groupby('course_category').size().to_frame('Amount of Courses per Category')
amount_courses

plt.bar(amount_courses.index, amount_courses['Amount of Courses per Category'])
plt.title('Amount of Courses per Category')
plt.tight_layout()
plt.ylabel('Amount')
plt.show()
```
